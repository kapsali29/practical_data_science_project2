```python
from yahoo_finance import Share
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.graphics as sgr
import matplotlib.pyplot as plt
import numpy as np
import time
from matplotlib.collections import LineCollection
from statsmodels.stats.outliers_influence import summary_table
from pandas_datareader import data as wb
from pprint import pprint
```

```python
firms = ['BK','BFED','BLK','CARV','C','CME','EVR','FII','FIG','GBL','LAZ','NTRS','BPOP','PFS','STT','BX','NDAQ']
```

First we find the tickers for the firms that we will investigate. To calculate the returns for each firm between 23-10-2007 and 20-10-2008 we must find the open  stock value and the close stock value for each day. Now we use the pandas_datareader package to acces yahoo finance to get the necassary information. Making use of DataReader method we have as output this information as pandas dataframe, the use of this function is similar to get_historical function but it returns a dataframe with index the dates that takes as arguments this function instead of list of dictionaries that returns function get_historical
We compute the daily returns for each firm using the following formula

\begin{equation*}
returns  = (Closing\;stock\;price - Opening\;stock\;price)/(Opening\;stock\;price)
\end{equation*}

We must note that as Open we mean Opening stock price and as Close we mean the Closing stock price. The function get_historical returns a list of dictionaries, each dictionaty contains the necassary information about each day, which was in days between the date set that function get_historical takes them as arguments.

```python
info = pd.DataFrame()
for firm in firms:
    temp = wb.DataReader(firm, 'yahoo', '2007-10-23','2008-10-20')
    firm_df = temp[['Open','Close']]
    firm_df['Open'] = pd.to_numeric(firm_df['Open'])
    firm_df['Close'] = pd.to_numeric(firm_df['Close'])
    string = firm +'_'+'returns'
    firm_df[string] = ((firm_df['Close'] - firm_df['Open'])/firm_df['Open'])*100
    firm_df = firm_df[string]
    info = pd.concat([info,firm_df], axis = 1,)
pprint(info)
```

```python
hist_data = wb.DataReader('^GSPC', 'yahoo', '2007-10-23', '2008-10-20')
sp500_df = hist_data[['Open','Close']]
sp500_df['Open'] = pd.to_numeric(sp500_df['Open'])
sp500_df['Close'] = pd.to_numeric(sp500_df['Close'])
sp500_df['returns'] = ((sp500_df['Close'] - sp500_df['Open'])/sp500_df['Open'])*100
info['sp'] = sp500_df['returns']
pprint(info)
```

Now we want to calculate the following formula
\begin{equation*}
{AR}_{it} = R_{it} - [\hat{\alpha_i} + \hat{\beta_i}R_{mt}]
\end{equation*}
First we have to estimate the $a_i$ and $b_i$. To do that we implement the folllowing linear regresion model, which is given by Question 2,

\begin{equation*} R_{it} = \alpha_i + \beta_iR_{mt} +\epsilon_{it}\end{equation*}

Assuming dependence between variales $R_{it}$ and $R_{mt}$, using the above linear regression model we will estimate the values for $a_i$ and $b_i$. So in the code below we create a model for returns of each firm and S&P 500 index returns. The parameters that occurs are the requested $a_i$ and $b_i$.

```python
a = []
b = []
for firm in firms:
    string = firm +'_'+'returns' + ' ~ sp'
    mod = smf.ols(string, data = info)
    res = mod.fit()
    a.append(res.params[0])
    b.append(res.params[1])
```

After that we add one column in our dataframe. This column is the S&P 500 index for this 10-day interval.

```python
market = Share('^GSPC')
temp = market.get_historical('2008-11-21','2008-12-08')
market_df = pd.DataFrame(temp)
market_df = market_df[['Date','Open','Close']]
market_df = market_df.set_index('Date')
market_df['Open'] = pd.to_numeric(market_df['Open'])
market_df['Close'] = pd.to_numeric(market_df['Close'])
market_df['returns'] = ((market_df['Close'] - market_df['Open'])/market_df['Open'])*100
R['sp'] = market_df['returns']
pprint(R)
```

Then having the dtaframe we compute the requested formula, for each firm, and we store the result to a dataframe

```python
i = 0
sec_df = pd.DataFrame()
for firm in firms:
    string = 'returns'+'_'+firm
    AR = R[string] - (a[i] + b[i]*R['sp'])
    string2 = 'AR'+'_'+firm
    sec_df[string2] = AR
    i = i+1
pprint(sec_df)
    
```

After that we want, now to calculate this formula, which give to us the cumulative abnormal returns
\begin{equation*} {CAR}[0, n]_i = \sum_{t=0}^{n}\mathrm{AR}_{it} \end{equation*}

We have store $AR_it$ to dataframe sec_df and we want to calculate the above formula from date 2008-11-21 to 2008-12-08. The following code calculates the requested returns ,CAR, for each company,we store the result to CAR dataframe

```python
event_days = ['2008-11-21','2008-11-24','2008-11-25','2008-11-26','2008-11-28','2008-12-01','2008-12-02','2008-12-03','2008-12-04','2008-12-05','2008-12-08']
```

```python
CAR = pd.DataFrame(index=sec_df.index)
for firm in firms:
    string5 = 'AR'+'_'+firm
    temp_list = []
    for i in range(len(event_days)):
        k = i +1
        temp_list.append(sum(sec_df[string5][event_days[:k]]))
    string7 = 'CAR'+'_'+firm
    CAR[string7] = temp_list
    
    
```

```python
pprint(CAR)
```

Next we add to dataframe to columns, these columns contains categorical variables, the first column is the schedule connections and the second column is personal connections, firms that don't have connections appended to zero 

```python
connections = pd.DataFrame()
connections['car0'] = CAR.loc['2008-11-21']
connections['car01'] = CAR.loc['2008-11-24']
connections['car010'] = CAR.loc['2008-12-08']
connections = connections.set_index([firms])
connections['schedule'] = [7,1,13,0,0,2,1,1,0,0,1,1,0,2,1,6,2]
connections['personal'] = [0,0,2,1,2,0,0,0,1,1,0,0,1,0,0,4,1]
pprint(connections)
```

Finally we implement the given linear regression models for $CAR[0,0]$, $CAR[0,1]$, $CAR[0,10]$ to find the dependence between between those variables above with Schedule connections and Personnal connections 

```python
a_sch = []
b_sch = []
a_pers = []
b_pers = []
cars = ['car0','car01','car010']
for car in cars:
    pprint(car)
    pprint('#######Schedule Connections ##############')
    string1 = car + ' ~ schedule'
    mod = smf.ols(string1, data = connections)
    res = mod.fit()
    pprint(res.params)
    a_sch.append(res.params[0])
    b_sch.append(res.params[1])
    pprint('#######Personal Connections ##############')
    string2 = car + ' ~ personal'
    mod1 = smf.ols(string2, data = connections)
    res1 = mod1.fit()
    pprint(res1.params)
    a_pers.append(res1.params[0])
    b_pers.append(res1.params[1])
    
```
